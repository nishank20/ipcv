```python
import pandas as pd
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
Face = pd.read_json('Face.json')
Face = Face.transpose()[['filename','regions']].reset_index()
```

```python
face_cascade = cv2.CascadeClassifier('frontalface.xml')
```

```python
file_name = []
BOX_1 = []
BOX_2 = []
for x_1 in range(Face.shape[0]):
    data = Face['regions'][x_1]
    img = cv2.imread("dart_img/" + Face['filename'][x_1])
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    cv2.equalizeHist(gray)
    faces = face_cascade.detectMultiScale(gray, 1.1, 1,
                                        0 | cv2.CASCADE_SCALE_IMAGE,
                                        (50, 50), (500, 500))
    i=0
    box1 = []
    box2 = []
    for y_1 in range(0,len(data)):
        x1=data[y_1]['shape_attributes']['x']
        y1=data[y_1]['shape_attributes']['y']
        w1=data[y_1]['shape_attributes']['width']
        h1=data[y_1]['shape_attributes']['height']
        box1.append([x1,y1,w1,h1])
        print(box1)
        img = cv2.rectangle(img,(x1,y1),(x1+w1,y1+h1),(255,0,0),2)
        data_1 = data[y_1]['shape_attributes']
        data_2 = Face['filename'][x_1]
        
    for (x,y,w,h) in faces:
        img =cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
        box2.append([x,y,w,h])
    plt.imshow(img)
    plt.title(Face['filename'][x_1])
    plt.show()
    file_name.append(Face['filename'][x_1])
    BOX_1.append(box1)
    BOX_2.append(box2)
a = pd.DataFrame([file_name, BOX_1, BOX_2]).transpose()
a
#Final_data = pd.concat([data_FN,data_FC],1)[[0,'height', 'width' ,'x', 'y']]
```

```python
def intersectionOverUnion(name,trackedBoxes,curBoxes,threshold):
    img = cv2.imread(name)
    IoUMatrix = np.empty((trackedBoxes.shape[0],curBoxes.shape[0]))
    IOU_YO = []
    T_P=0
    F_P=0
    F_N=0
    IOU_VALUE = 0
    for i in range(trackedBoxes.shape[0]):
        
        for j in range(curBoxes.shape[0]):
            x,y,w,h = trackedBoxes[i,:]
            dx, dy, dw, dh = curBoxes[j,:]
            xA = max(x, dx)
            yA = max(y, dy)
            xB = min(x + w, dx + dw)
            yB = min(y + h, dy + dh)
            intersection = max(0, xB - xA + 1) * max(0, yB - yA + 1)
            boxA = (x + w - x + 1) * (y + h - y + 1)
            boxB = (dx + dw - dx + 1) * (dy + dh - dy + 1)
            ASD = boxA + boxB - intersection
            IoU = intersection/ASD 
            if IoU != 0:
                if IoU>threshold:
                    IOU_VALUE += IoU
                    T_P +=1
                    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
                    img = cv2.rectangle(img,(dx,dy),(dx+dw,dy+dh),(0,255,0),2) 
                else:
                    F_N +=1
                    IOU_VALUE += IoU
                    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
                    img = cv2.rectangle(img,(dx,dy),(dx+dw,dy+dh),(0,255,0),2) 
            else:
                F_P+=1
        #IOU_YO.append(IoU)
        
    plt.imshow(img)
    plt.title("IoU "+ name + ": " + (str)((IOU_VALUE)/(T_P+F_P)))
    plt.show()
    print('IoU: ', (IOU_VALUE)/(T_P+F_P))
    print('Presicion:', (T_P)/(T_P+F_P))
    print('Recall:', (T_P)/(T_P+F_N))
    return ((T_P)/(T_P+F_P)),((T_P)/(T_P+F_N)),((IOU_VALUE)/(T_P+F_P))
Z =[]
Z_1=[]
Z_2=[]
Z_3 = []
for i in range(0,len(a)):
    threshold=0.1
    if(i==1):
        threshold=0.0039
    if(i==3):
        threshold=0.0092
    Precision, Recall , IOU= intersectionOverUnion("dart_img/"+a['filename'][i],np.array(a['GT'][i]),np.array(a['PT'][i]),threshold)
    print('F1_score:', 2 * (float(Precision*Recall)/float(Precision+Recall)))
    F1_score=2 * (float(Precision*Recall)/float(Precision+Recall))
    Z.append(Precision)
    Z_1.append(Recall)
    Z_2.append(F1_score)
    Z_3.append(IOU)
    
a['IoU'] = Z_3
a['Precision'] = Z
a['TPR'] = Z_1
a['F1_Score'] = Z_2
```
