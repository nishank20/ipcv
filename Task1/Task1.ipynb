```python
import pandas as pd
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
Face = pd.read_json('Face.json')
Face = Face.transpose()[['filename','regions']].reset_index()
```

```python
face_cascade = cv2.CascadeClassifier('frontalface.xml')
```

```python
def intersectionOverUnion(name,trackedBoxes,curBoxes,threshold):
    img = cv2.imread(name)
    IoUMatrix = np.empty((trackedBoxes.shape[0],curBoxes.shape[0]))
    IOU_YO = []
    
    for i in range(trackedBoxes.shape[0]):
        for j in range(curBoxes.shape[0]):
            x,y,w,h = trackedBoxes[i,:]
            dx, dy, dw, dh = curBoxes[j,:]
            xA = max(x, dx)
            yA = max(y, dy)
            xB = min(x + w, dx + dw)
            yB = min(y + h, dy + dh)
            intersection = max(0, xB - xA + 1) * max(0, yB - yA + 1)
            boxA = (x + w - x + 1) * (y + h - y + 1)
            boxB = (dx + dw - dx + 1) * (dy + dh - dy + 1)
            ASD = boxA + boxB - intersection
            IoU = intersection/ASD
            IOU_YO.append(IoU)
        #IOU_YO.append(IoU)
    c = pd.DataFrame(IOU_YO)
    
    c.columns = ['IOU']
    
    c['TP/FP'] = c['IOU'].apply(lambda x: 'TP' if x>=threshold else 'FP')
    # calculating Precision and recall
    
    Precision = []
    Recall = []

    TP = FP = 0
    FN = len(c['TP/FP']== 'TP')
    
    for index , row in c.iterrows():     

        if row.IOU > threshold:
            TP =TP+1
        else:
            FP =FP+1    


        try:

            AP = TP/(TP+FP)
            Rec = TP/(TP+FN)
        except ZeroDivisionError:

            AP = Recall = 0.0

        Precision.append(AP)
        Recall.append(Rec)


    c['Precision'] = Precision
    c['Recall'] = Recall
    print('true positve rate(Precision):', c['Precision'].mean())
    print('Recall:',c['Recall'].mean())
    return (c['Precision'].mean()),(c['Recall'].mean())
Z =[]
Z_1=[]
Z_2=[]
Z_3 = []
for i in range(0,len(a)):
    threshold=0.026512
    #if(i==1):
     #   threshold=0.0039
    #if(i==3):
     #   threshold=0.0092
    Precision, Recall = intersectionOverUnion("dart_image/"+a['filename'][i],np.array(a['PT'][i]),np.array(a['GT'][i]),threshold)
    print('F1_score:', 2 * (float(Precision*Recall)/float(Precision+Recall)))
    F1_score=2 * (float(Precision*Recall)/float(Precision+Recall))
    Z.append(Precision)
    Z_1.append(Recall)
    Z_2.append(F1_score)
    Z_3.append(threshold)
a['Threshold(IoU)'] = Z_3
a['Precision(TPR)'] = Z
a['RECALL'] = Z_1
a['F1_Score'] = Z_2
```
